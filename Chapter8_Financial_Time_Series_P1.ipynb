# Rolling statostics

```python
# check chat gpt description for rollign statoitsics its quite cool to spt trends in finanical time series data sequences
```

```python
data.head()
```

```python
#this works with a single instrument only
```

```python
sym = 'AAPL.O'
data_2 = pd.DataFrame(data[sym]).dropna()
```

```python
data_2.tail()
```

```python
window = 20
```

```python
data_2['min'] = data_2[sym].rolling(window=window).min()
```

```python
data_2.head(50)
```

```python
data.head()
```

```python
data_2.head()
```

```python
data_2.dropna().head(100)
```

```python
data_2['mean'] = data_2[sym].rolling(window=window).mean()
data_2['std'] = data_2[sym].rolling(window=window).std()
data_2['median'] = data_2[sym].rolling(window=window).median()
data_2['max'] = data_2[sym].rolling(window=window).max()
data_2['ewma'] = data_2[sym].ewm(halflife=0.5, min_periods=window).mean()
```

```python
data_2.dropna().head(100)
```

```python
ax = data_2[['min', 'mean', 'max']].iloc[-200:].plot(
    figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8)
```
